```python
data_older_population = pd.read_csv('C:\\playdata\\23_02_01_python_mini_project\\data_older_population.csv')
data_older_population = data_older_population.astype({'소계':'int'})
data_older_population = data_older_population.sort_values('동별', ascending=False)
```

```python
data_older_accident_number = pd.read_csv('C:\\playdata\\23_02_01_python_mini_project\\ped.csv')
data_older_accident_number = data_older_accident_number.sort_values('지역구', ascending=False)
```

```python
# 지역구별 2022년 서울시 노인 인구
plt.bar(data_older_population["동별"], data_older_population["소계"], color = 'b')
plt.xticks(rotation=45)
plt.title("22년도 서울시 노인 인구")
plt.ylabel("인구 수")
plt.xlabel("지역구")
plt.show()
```

```python
# 지역구별 2022 서울시 노인 사고 건수
plt.bar(data_older_accident_number["지역구"], data_older_accident_number["발생건수"], color = 'b')
plt.xticks(rotation=45)
plt.title("22년도 서울시 노인 사고 수")
plt.ylabel("사고 수")
plt.xlabel("지역구")
plt.show()
```

```python
# column에 문제가 있던 코드
fig, ax3 = plt.subplots()

bar_width = 0.35
alpha = 0.5
plt.figure(figsize = (15,10))
N = len(data_older_population['동별'].unique())
index = np.arange(N)
p3 = ax3.bar(data_older_population['동별'], data_older_population["소계"], alpha=alpha, color='b', width=0.5)

ax4 = ax3.twinx()
p4 = ax4.bar(index + bar_width, data_older_accident_number['발생건수'], alpha=alpha, color='r', width=0.5)

ax3.legend((p3, p4), ('노인 인구 수', '노인 교통사고 발생 건수'), fontsize = 12, loc='upper right')
ax3.set_title("얼마나 많은 노인이 사고를 겪는가")

plt.show()
```

## 최근 5년간 보행자 교통사고 사망자 비교

```python
difference_total_to_old = pd.DataFrame({"17년": [1675, 906], "18년": [1487, 842], 
                                        "19년": [1302, 743], "20년": [1093, 628], "21년": [1018, 601]}, index= ["전체 사망자", "노인 사망자"])
difference_total_to_old
```

```python
difference_total_to_old.loc["노인 사망비율"] = difference_total_to_old.loc["노인 사망자"] / difference_total_to_old.loc["전체 사망자"] * 100
difference_total_to_old
```

```python
plt.plot(difference_total_to_old.loc["노인 사망비율"])
```

```python
fig, ax1 = plt.subplots()

bar_width = 0.35
alpha = 0.5
plt.figure(figsize = (10,10))
N = len(difference_total_to_old.columns.unique())
index = np.arange(N)
p1 = ax1.bar(difference_total_to_old.columns, difference_total_to_old.loc['전체 사망자'], bar_width, alpha=alpha)
p2 = ax1.bar(index + bar_width, difference_total_to_old.loc["노인 사망자"], bar_width, alpha=alpha)

ax1.legend((p1, p2), ('전체 사망자 수', '노인 사망자 수'), fontsize = 12, loc='upper right')

ax2= ax1.twinx()
p3 = ax2.plot(difference_total_to_old.columns, difference_total_to_old.loc["노인 사망비율"], label="Percentage")
ax2.legend(loc='upper left', fontsize= 12)
ax1.set_title("늘어가는 고령자 교통사고 사망자")
plt.show()
```
